```python
import pandas as pd

# Initialize lists to hold DrugBank IDs, UniProt IDs, protein sequences, and interaction labels
drugbank_ids = []
uniprot_ids = []
protein_sequences = []
interaction_labels = []

# Parse the .txt file to extract DrugBank IDs, UniProt IDs, protein sequences, and interaction labels
with open('/content/DrugBank.txt', 'r') as file:  # Replace with the actual path to your .txt file
    for line in file:
        line = line.strip()

        # Split each line by whitespace to get DrugBank ID, UniProt ID, Sequence, and Label
        parts = line.split()

        # Ensure the line has all required parts (DrugBank ID, UniProt ID, Protein Sequence, Interaction Label)
        if len(parts) >= 4:
            drugbank_id = parts[0]
            uniprot_id = parts[1]
            protein_sequence = ''.join(parts[2:-1])  # Join all parts in between as the protein sequence
            interaction_label = int(parts[-1])  # The last part is the interaction label

            # Append data to respective lists
            drugbank_ids.append(drugbank_id)
            uniprot_ids.append(uniprot_id)
            protein_sequences.append(protein_sequence)
            interaction_labels.append(interaction_label)

# Create a DataFrame from the parsed data
data = {
    'drugbank_id': drugbank_ids,
    'uniprot_id': uniprot_ids,
    'protein_sequence': protein_sequences,
    'interaction_label': interaction_labels
}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
csv_path = '/mnt/data/drugbank_protein_data_with_interaction.csv'
df.to_csv(csv_path, index=False)

# Display the first few rows of the DataFrame
df.head(), csv_path
```

```python
# CONVERTING SMILE SEQUENCES INTO MOLECULAR GRAPHS
from rdkit import Chem
import networkx as nx
from rdkit.Chem import rdmolops


def smiles_to_graph(smiles):
    # Convert the SMILES string to an RDKit molecule
    mol = Chem.MolFromSmiles(smiles)

    if mol is None:
        raise ValueError("Invalid SMILES string")

    # Create an empty graph
    graph = nx.Graph()

    # Add nodes with atom properties as features
    for atom in mol.GetAtoms():
        graph.add_node(atom.GetIdx(), symbol=atom.GetSymbol(), atomic_num=atom.GetAtomicNum())

    # Add edges with bond type as features
    for bond in mol.GetBonds():
        graph.add_edge(
            bond.GetBeginAtomIdx(),
            bond.GetEndAtomIdx(),
            bond_type=bond.GetBondTypeAsDouble()  # Represent bond types as doubles (1.0 for single, 2.0 for double, etc.)
        )

    return graph

# Iterate over each SMILES string in the 'smiles' column
graphs = [] # List to store individual molecular graphs
for smile in df['smiles']:  # Iterate over each SMILES string
    try:
        graph = smiles_to_graph(smile) # Convert the current SMILES string to a graph
        graphs.append(graph) # Add the current graph to the list of graphs
    except ValueError:
        print(f"Skipping invalid SMILES: {smile}") # Print a message for invalid SMILES and continue to the next

# Now 'graphs' is a list containing a graph for each valid SMILES string
print("Graphs created for each SMILES string.")
```

```python
# FEATURE EXTRACTION
import torch
from rdkit import Chem
from rdkit.Chem import rdchem

def extract_node_features(atom):
    # Example: atomic number, atom degree, hybridization, aromaticity
    features = [
        atom.GetAtomicNum(),
        atom.GetDegree(),
        atom.GetHybridization(),
        atom.GetIsAromatic()
    ]
    return torch.tensor(features, dtype=torch.float)

def extract_edge_features(bond):
    # Example: bond type (single, double, etc.), aromaticity
    bond_type = bond.GetBondType()
    features = [
        bond_type == rdchem.BondType.SINGLE,
        bond_type == rdchem.BondType.DOUBLE,
        bond_type == rdchem.BondType.TRIPLE,
        bond.GetIsAromatic()
    ]
    return torch.tensor(features, dtype=torch.float)

# Assuming 'df' and 'smiles_to_graph' are defined from previous cells
# and 'df' has a 'smiles' column
# Select a specific SMILES string to process, or loop through all of them
smiles_string = df['smiles'].iloc[0]  # Example: using the first SMILES string
mol = Chem.MolFromSmiles(smiles_string)  # Create the RDKit molecule

if mol is not None:
    # Collect node and edge features
    node_features = [extract_node_features(atom) for atom in mol.GetAtoms()]
    edge_features = [extract_edge_features(bond) for bond in mol.GetBonds()]

    print("Node Features:", node_features)
    print("Edge Features:", edge_features)
else:
    print(f"Invalid SMILES string: {smiles_string}")
```

```python
import torch
import torch.nn as nn
from torch_geometric.nn import TransformerConv
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score, matthews_corrcoef, confusion_matrix

class GraphTransformer(nn.Module):
    def __init__(self, in_channels, out_channels, heads=8):
        super(GraphTransformer, self).__init__()
        # Multi-head self-attention for message passing
        self.transformer = TransformerConv(in_channels, out_channels, heads=heads, concat=True)

    def forward(self, x, edge_index):
        # x: node features, edge_index: adjacency list of the graph
        return self.transformer(x, edge_index)

    def evaluate(self, y_true, y_pred, task_type='binary'):
        # Assuming y_true and y_pred are numpy arrays or tensors containing true and predicted labels
        y_true = y_true.cpu().numpy() if torch.is_tensor(y_true) else y_true

        # Convert y_pred to 1D binary labels or multiclass labels
        if task_type == 'binary':
            y_pred = (y_pred > 0.5).long().cpu().numpy()
        else:
            y_pred = y_pred.argmax(dim=1).cpu().numpy()  # For multiclass classification

        accuracy = accuracy_score(y_true, y_pred)
        precision = precision_score(y_true, y_pred, average='binary' if task_type == 'binary' else 'weighted')
        recall = recall_score(y_true, y_pred, average='binary' if task_type == 'binary' else 'weighted')
        f1 = f1_score(y_true, y_pred, average='binary' if task_type == 'binary' else 'weighted')

        # For AUC, use 'macro' average for multiclass
        if task_type == 'binary':
            auc = roc_auc_score(y_true, y_pred)
            aupr = average_precision_score(y_true, y_pred)
        else:
            auc = roc_auc_score(y_true, y_pred, average='macro', multi_class='ovr')
            aupr = average_precision_score(y_true, y_pred, average='macro')

        mcc = matthews_corrcoef(y_true, y_pred)
        cm = confusion_matrix(y_true, y_pred)

        return {
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1,
            'AUC': auc,
            'AUPR': aupr,
            'MCC': mcc,
            'Confusion Matrix': cm
        }

# Example instantiation
model = GraphTransformer(in_channels=4, out_channels=8)  # Adjust in/out channels as needed
print(model)

# Assuming you have data in the format (node_features, edge_index, true_labels)
# Example forward pass and evaluation
x = torch.randn(10, 4)  # 10 nodes with 4 features each (sample data)
edge_index = torch.tensor([[0, 1, 2], [1, 2, 3]])  # Example edge list
true_labels = torch.randint(0, 2, (10,))  # Random binary labels for 10 nodes
output = model(x, edge_index)

# For simplicity, assume we predict based on some threshold
y_pred = (output > 0.5).float()  # Binary classification prediction (example)

# Evaluation
metrics = model.evaluate(true_labels, y_pred, task_type='multiclass')  # Set task_type based on your problem
print(metrics)
```

```python
import pandas as pd


df_smiles = pd.read_csv('/content/drugbank_smiles.csv')
df_smiles.head()
```

```python
df_protein = pd.read_csv('/content/drugbank_protein_data.csv')
df_protein.rename(columns={'pubmed_id': 'uniprot_id'}, inplace=True)
df_protein.head()
```

```python
df = pd.merge(df_smiles, df_protein, on='drugbank_id', how='left')
df.head()
```

```python
columns_to_drop = ['cas', 'logP ALOGPS', 'logP ChemAxon', 'solubility ALOGPS', 'pKa (strongest basic)',
                   'pKa (strongest acidic)']
df = df.drop(columns=columns_to_drop)
# dropping rows where protein sequece is null
df = df.dropna(subset=['sequence'])
df.head()
```
